## Investigating time-to-human-response

```python
# Does the distribution change before/after Nov 3rd, which is DST change day?
pre_nov3 = C[C['first_msg'] < pd.to_datetime('2013-11-03')]
post_nov3 = C[C['first_msg'] >= pd.to_datetime('2013-11-03')]

f, (a1, a2) = plt.subplots(1, 2, figsize=(8, 4))
plot_durations(pre_nov3['duration_to_first_resp'], tmin=-1*60*60, tmax=15*60, title="Events before Nov 3", label="Time from first message to response (minutes)", ax=a1)
plot_durations(post_nov3['duration_to_first_resp'], tmin=-1*60*60, tmax=15*60, title="Events after Nov 3", label="Time from first message to response (minutes)", ax=a2)
```

Plot distribution of time to human response; do the different conversation rating classes correlate with different time-to-first-response distributions?

```python
f, a = plt.subplots()
for g, vs in C.groupby('conv_rating')['duration_to_first_resp_nonneg']:
    vs = vs[vs < 0.25 * 60 * 60 * 10**9]
    #vs = vs[vs < float(np.timedelta64(1.5, 'h'))]
    vs.dropna().plot(ax=a, kind='kde', label=g)
a.legend()
a.set_xticklabels(map(lambda x: nano_to_sec(x) / 60, a.get_xticks()))
a.set_xlabel('Time to human response (minutes)')
a.set_title('Human response time distributions\nfor different teen outcomes');
```

```python
from analysis.processing import pctile_of, group_summary
```

Do conversations with first-response times under 10 minutes lead to better mean ratings?

```python
dur_pos = C[C['duration_to_first_resp_nonneg'] > 0]
dur_pos = dur_pos[dur_pos['conv_rating'].notnull()]

dur_pos['duration_to_first_resp_nonneg_decile'] = (dur_pos['duration_to_first_resp_nonneg'].dropna() < 10 * 60 * 10**9)

mean_by_decile_resptime = dur_pos.groupby('duration_to_first_resp_nonneg_decile').apply(group_summary)
mean_by_decile_resptime
```

How about conversations that are specifically about suicide — does first-response time matter?

```python
dur_pos_suicide = C[((C['duration_to_first_resp_nonneg'] > 0) * (C.is_suicide_issue))]
dur_pos_suicide = dur_pos_suicide[dur_pos_suicide['conv_rating'].notnull()]
dur_pos_suicide['duration_to_first_resp_nonneg_decile'] = pctile_of(dur_pos_suicide['duration_to_first_resp_nonneg'])
dur_pos_suicide['duration_to_first_resp_nonneg_decile'] = (dur_pos_suicide['duration_to_first_resp_nonneg'].dropna() < 10 * 60 * 10**9)
mean_by_decile_resptime_suicide = dur_pos_suicide.groupby('duration_to_first_resp_nonneg_decile').apply(group_summary)
mean_by_decile_resptime_suicide
```
